# 以下将对Airbnb2019建模
## 使用数据为数据预处理后生成数据
## 会尽量将可复用方法前移

```python
#调包
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
from IPython.display import display
import matplotlib.pyplot as plt
```

```python
training_set = pd.read_csv('./train.csv')
display(training_set)
test = pd.read_csv('./test.csv')
display(test)
```

```python
#第一列为target, 剩余为feature
y_train = training_set['price']
x_train = training_set.drop(['price'],axis=1)
print(len(y_train),len(x_train))
#对于new_host的test
y_test = test['price']
x_test = test.drop(['price'],axis=1)
```

```python
#针对old_host划分训练集&测试集，本测试集用于评估模型准确程度
#至于new_host 如果出现的predict与实际偏差越大，越能说明我们模型和建议的必要性
from sklearn.cross_validation import train_test_split
oh_x_train,oh_x_test,oh_y_train,oh_y_test = train_test_split(x_train,y_train,train_size=0.8)
print(len(oh_x_train),len(oh_x_test),len(oh_y_train),len(oh_y_test))
```

# 线性回归表现太差了………………

```python
#线性回归
from sklearn import linear_model
from sklearn.metrics import mean_squared_error,r2_score
lin_reg = linear_model.LinearRegression()
lin_reg.fit(oh_x_train,oh_y_train)
lin_reg_pred = lin_reg.predict(oh_x_test)

print('Coefficients: \n', lin_reg.coef_)
print('Mean squared error: %.2f'
      % mean_squared_error(oh_y_test, lin_reg_pred))
print('Coefficient of determination: %.2f'
      % r2_score(oh_y_test, lin_reg_pred))
```

虽然同为线性模型，但是Logistics表现优于LinearRegression

```python
#Logistics Regression
from sklearn.linear_model import LogisticRegression
log_reg = linear_model.LogisticRegression()
log_reg.fit(oh_x_train,oh_y_train)
log_reg_pred = log_reg.predict(oh_x_test)

print('Coefficients: \n', log_reg.coef_)
print('Mean squared error: %.2f'
      % mean_squared_error(oh_y_test, log_reg_pred))
print('Coefficient of determination: %.2f'
      % r2_score(oh_y_test, log_reg_pred))
```

```python
#SVR
from sklearn import svm
svr_reg = svm.SVR(kernel='linear', C=1.25)
svr_reg.fit(oh_x_train,oh_y_train)
svr_reg_pred = svr_reg.predict(oh_x_test)

print('Coefficients: \n', svr_reg.coef_)
print('Mean squared error: %.2f'
      % mean_squared_error(oh_y_test, svr_reg_pred))
print('Coefficient of determination: %.2f'
      % r2_score(oh_y_test, svr_reg_pred))
```

```python
#SVR
from sklearn import svm
svr_reg = svm.SVR(kernel='rbf', C=100, gamma=0.1, epsilon=.1)
svr_reg.fit(oh_x_train,oh_y_train)
svr_reg_pred = svr_reg.predict(oh_x_test)

print('Mean squared error: %.2f'
      % mean_squared_error(oh_y_test, svr_reg_pred))
print('Coefficient of determination: %.2f'
      % r2_score(oh_y_test, svr_reg_pred))
```

```python
#SVR
from sklearn import svm
svr_reg = svm.SVR(kernel='poly', C=100, gamma='auto', degree=3, epsilon=.1,coef0=1)
svr_reg.fit(oh_x_train,oh_y_train)
svr_reg_pred = svr_reg.predict(oh_x_test)

print('Mean squared error: %.2f'
      % mean_squared_error(oh_y_test, svr_reg_pred))
print('Coefficient of determination: %.2f'
      % r2_score(oh_y_test, svr_reg_pred))
```

```python
#NN
from sklearn.neural_network import MLPRegressor
regr = MLPRegressor(random_state=1, max_iter=500).fit(oh_x_train,oh_y_train)
nn_reg_pred = regr.predict(oh_x_test)
print('Mean squared error: %.2f'
      % mean_squared_error(oh_y_test, nn_reg_pred))
print('Coefficient of determination: %.2f'
      % r2_score(oh_y_test, nn_reg_pred))
```

```python
from sklearn.linear_model import Ridge,RidgeCV
Lambdas = np.logspace(-5,2,200)
rigdeCV = RidgeCV(alphas=Lambdas,normalize=True,scoring='neg_mean_squared_error',cv=15)

rigdeCV.fit(oh_x_train,oh_y_train)
predictedResult = rigdeCV.predict(oh_x_test)
print('Mean squared error: %.2f'
      % mean_squared_error(oh_y_test, predictedResult))
print('Coefficient of determination: %.2f'
      % r2_score(oh_y_test, predictedResult))
```

```python
from sklearn.linear_model import ElasticNet

enet = ElasticNet(alpha=0.1, l1_ratio=0.7)

enet.fit(oh_x_train,oh_y_train)
y_pred_enet = enet.predict(oh_x_test)
print('Mean squared error: %.2f'
      % mean_squared_error(oh_y_test, y_pred_enet))
print('Coefficient of determination: %.2f'
      % r2_score(oh_y_test, y_pred_enet))
```

```python
#DT
from sklearn.tree import DecisionTreeRegressor
regr_1 = DecisionTreeRegressor(max_depth=2)
regr_2 = DecisionTreeRegressor(max_depth=5)
regr_1.fit(oh_x_train,oh_y_train)
regr_2.fit(oh_x_train,oh_y_train)
y_1 = regr_1.predict(oh_x_test)
y_2 = regr_2.predict(oh_x_test)

print('Mean squared error-y_1: %.2f'
      % mean_squared_error(oh_y_test, y_1))
print('Coefficient of determination-y_1: %.2f'
      % r2_score(oh_y_test, y_1))
print('Mean squared error-y_2: %.2f'
      % mean_squared_error(oh_y_test, y_1))
print('Coefficient of determination-y_2: %.2f'
      % r2_score(oh_y_test, y_2))
```

```python
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor() #default 5
neigh.fit(oh_x_train,oh_y_train)
y_nei_pred = neigh.predict(oh_x_test)
print('Mean squared error: %.2f'
      % mean_squared_error(oh_y_test, y_nei_pred))
print('Coefficient of determination: %.2f'
      % r2_score(oh_y_test, y_nei_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
RF = RandomForestRegressor()
RF.fit(oh_x_train,oh_y_train)
y_rf_pred = RF.predict(oh_x_test)
print('Mean squared error: %.2f'
      % mean_squared_error(oh_y_test, y_rf_pred))
print('Coefficient of determination: %.2f'
      % r2_score(oh_y_test, y_rf_pred))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
est = GradientBoostingRegressor()
est.fit(oh_x_train,oh_y_train)
y_est_pred = est.predict(oh_x_test)
print('Mean squared error: %.2f'
      % mean_squared_error(oh_y_test, y_est_pred))
print('Coefficient of determination: %.2f'
      % r2_score(oh_y_test, y_est_pred))
```
